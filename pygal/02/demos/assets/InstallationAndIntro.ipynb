#### Let's import the pygal and initialize the Pygal object. 

```python
import pygal
import os
import webbrowser
```

```python
bar_chart = pygal.Bar()
```

Take some data to show the bar. Figures are imaginary 

```python
bar_chart.title = "Pet Population By Animal Per 100 households"
```

Add the data to the bar chart.

```python
bar_chart.add('Dogs', 41)
```

```python
bar_chart = pygal.Bar(height = 360, 
                      width = 640, 
                      explicit_size=True)
```

```python
bar_chart.title = "Pet Population By Animal Per 100 households"

bar_chart.add('Dogs', 41)
```

Add the data to the chart at one go.

```python
bar_chart.add('Cats', 20)
bar_chart.add('Bird', 12)
bar_chart.add('Horses', 7)
bar_chart.add('Exotic Pets', 2)
bar_chart.add('No Pets', 18)
```

```python
bar_chart.render_to_file('pet_bar_chart.svg')
```

**Import the os module and get the absolute path to be able to open in the browser.**

```python
def open_in_browser(file_to_open):    
    file_path = os.path.realpath(file_to_open)
    file_url = 'file://' + file_path
    # browser = webbrowser.get('chrome')
    browser = webbrowser.get() # WSL 2 needs default and wslu APT package to be installed
    browser.open(file_url)
```

```python
open_in_browser('pet_bar_chart.svg')
```

### Horizontal Bar
The Bar chart can be flipped as Horizontal BarChart.

```python
horizontal_bar = pygal.HorizontalBar(height = 360, 
                                     width = 640, 
                                     explicit_size=True)

horizontal_bar.title = "Pet Population By Animal Per 100 households"
```

```python
horizontal_bar.add('Dogs',41)
horizontal_bar.add('Bird',12)
horizontal_bar.add('Horses', 7)
horizontal_bar.add('Exotic Pets',2)
horizontal_bar.add('No Pets',18)
```

```python
horizontal_bar.render_to_file('pet_horizontal_bar_chart.svg')
```

```python
open_in_browser('pet_horizontal_bar_chart.svg')
```
